```python
import requests
import bs4

import pandas as pd
from pandas import DataFrame
```

```python
OLD_FILE = '../Data/playerData_avg.csv'

playDF = pd.read_csv(OLD_FILE)
```

```python
LOGIN_URL = 'https://r.espn.go.com/espn/memberservices/pc/login'
# Fill in your details here to be posted to the login form.
payload = {
    'username': 'filiwickers',
    'password': 'blueman',
    'submit': 'Sign In',
    'failedAttempts':2,
    'SUBMIT':1,
    'aff_code' : 'espn_fantgames',
    'cookieDomain' :'.go.com',
    'multipleDomains':'true',
    
}

FANTASY_URL = 'http://games.espn.go.com/fba/playertable/prebuilt/freeagency'

params = (
    ('leagueId',14029),
    ('teamId',1),
    ('seasonId',2015),
    ('','undefined'),
    ('avail',4), # This means on rosters
    ('scoringPeriodID',7), # &scoringPeriodId=7 for 2015 season data

    ('view','stats'),
    ('context','freeagency'),
    ('version', 'last30'), #or ('version','projections' , &version=currSeason, &version=last15)
    ('r',48197013),
    ('startIndex',0),
) 
#?leagueId=14029&teamId=1&seasonId=2015&=undefined&avail=-1&scoringPeriodId=1&view=stats&context=freeagency&version=lastSeason&r=48197013#

NUM_PLAYERS = 11 * 12 #11 in case of IR
startIndex = 0

players = []
columns = []
with requests.Session() as s:
    s.post(LOGIN_URL, data=payload)
 
    while startIndex < (NUM_PLAYERS):
        params = list(params)[:-1] + [('startIndex',startIndex)]
        params = tuple(params)

        r = s.get(FANTASY_URL, params=params)    
        print(r.url)
        
        soup = bs4.BeautifulSoup(r.text)
        players.extend(soup.find_all("tr", class_="pncPlayerRow"))
        startIndex += 50    
  
    columns = soup.find('tr', class_='playerTableBgRowSubhead')

len(players)
```

```python
cols = [col.text.strip() for col in columns.find_all('td')]
players_clean = [[col.text.strip() for col in player.find_all('td')] for player in players]
```

```python
owners = {}
i=0
while len(owners) < 12:
    owner_td = players[i].find_all('td')[2]
    owner_name = owner_td.text
    owner_id = owner_td.find('a')['href'].split('&')[1].split('=')[1]
    owners[owner_name] = owner_id
    i += 1
        
owners
```

```python
playerDF = DataFrame(players_clean, columns=cols)

def get_injured(player):
    player = player.split(u'\xa0')
    if player[-1] == 'O':
        print (player[0])
        return True
    return False

def clean_df(df):
    df['PLAYER'] = df['PLAYER, TEAM POS'].apply(lambda x: x.split(',')[0].strip())
    df['POS'] = df['PLAYER, TEAM POS'].apply(lambda x: x.split(u'\xa0')[1].split(','))
    df['POS'] = df['POS'].apply(lambda x: [item.strip() for item in x])
    df['TEAM'] = df['PLAYER, TEAM POS'].apply(lambda x: x.split(',')[1][:4].strip())
    df['INJURED'] = df['PLAYER, TEAM POS'].apply(get_injured)
    df['OWNER'] = df['TYPE']
    df = df[['PLAYER', 'TEAM','OWNER', 'POS', 'INJURED', 'PR15', 'MIN', '3PM', 'REB', 'PTS', 'STL', 'AST', 'BLK']]
    df = df.rename(columns={'MIN':'MIN_2015', '3PM':'3PM_2015', 'REB':'REB_2015', 'PTS':'PTS_2015', 
                            'STL':'STL_2015', 'AST':'AST_2015', 'BLK':'BLK_2015'})
    df['OWNER_ID'] = df['OWNER'].apply(lambda x: owners[x])

    df['PLAYER'] = df['PLAYER'].apply(lambda x: x.replace('*',''))
    df['PLAYER'] = df['PLAYER'].apply(lambda x: ' '.join(x.split()))
    df = df.replace(to_replace='--', value='NaN')

    return df

playerDF = clean_df(playerDF)

playerDF
```

```python
tempDF = playDF.filter(regex='PLAYER|_AVG')
playersFull = playerDF.merge(tempDF, on='PLAYER', how='left')
playersFull['TEAM'] = playersFull['TEAM'].apply(lambda x: str(x).upper())
```

```python
injured = DataFrame(playersFull[playersFull['INJURED']]['PLAYER']
, columns=['PLAYER','INJURED_START','INJURED_END'])

injured.to_csv('../data/playerOwner_inj.csv', index=False)
```
