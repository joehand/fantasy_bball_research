```python
""" Import the data. 

    We need three sets of data:
        - Player Database with stats, owners, positions, and injuries
        - Fantasy Schedule with team1 vs team2 (team IDs)
        - NBA Team schedule (according to fantasy weeks, e.g. combine all-star week into 1)
"""

PLAYER_FILE = '../Data/playerOwner_updated.csv'
FANTASY_SCHEDULE_FILE = '../Data/fantasy_schedule.csv'
NBA_SCHEDULE_FILE = '../Data/nbaFantasySched.csv'

players = pd.read_csv(PLAYER_FILE)
fantasy = pd.read_csv(FANTASY_SCHEDULE_FILE)
nba = pd.read_csv(NBA_SCHEDULE_FILE)

# Clean up some data
players['INJURED'] = players['INJURED'].fillna(False) 
del(players['MIN_AVG']) # we don't need this for now
players['POS'] = players['POS'].apply(lambda x: literal_eval(x)) # make POS a list not string
```

```python
""" This cell combines the fantasy schedule and NBA schedule into a list & dict for easy lookup
  
    We end up with this:
        weeks_data = [
                {
                    week: 1,
                    teams: {
                        'PDX' : 4,
                        'DEN' : 3,
                        ...
                    },
                    matches: [
                        {
                            team1_id: 1,
                            team1_owner: 'Joe',
                            team2_id: 2,
                            team2_owner: 'Fish',
                        },
                        ...
                    ]
                }
                ...
        ]
"""
weekly_data = []
NUM_WEEKS = 19

for week in range(1,NUM_WEEKS+1):
    teams = nba[nba['Week'] == week].to_dict(outtype='records')[0]
    teams.pop("Week")
    teams.pop("Date")
    
    matches = fantasy[fantasy['week'] == week][['team2_id', 'team2_owner', 'team1_id', 'team1_owner']].to_dict(outtype='records')
    
    item = {
        'week' : week,
        'teams' : teams,
        'matches' : matches
    }
    weekly_data.append(item)
```

```python
""" These are all the important calculations. 
"""

# Lineup possibilities. For G, F/C, UTIL, I pass a list off possible positions (e.g. G = ['PG','SG'])
LINEUP = [['PG'], ['SG'], ['SF'], ['PF'], ['C'],  ['PG', 'SG'], ['SF', 'PF', 'C'], ['PG', 'SG', 'SF', 'PF', 'C']]

def get_unique_lineups(team):
    """ Returns a list of each unique and valid lineup as DF
        
        Params:
        - team = subset of Players DataFrame for each team with POS column
        
        Returns:
        - List of Unique Lineup DataFrames (5-45 lineups)
    
    TODO: This is pretty slow. 
          This only needs to be run once on a team, just need to figure out injuries ahead of time
    """
    
    # Create a list of POS posibilities, e.g. player_positions = [['Rondo', 'Smart'], ['Kawhi', 'Affalo'], ...]
    player_positions = []
    for pos in LINEUP:
        player_pos = team[team['POS'].map(lambda x: any(i in pos for i in x))].index.values
        player_positions.append(player_pos.tolist())
    
    # Combine all lists into a set of unique lineups of length 8
    unique_lineups = set()
    for x in product(*player_positions):
        datum = frozenset(x)
        if len(datum) == 8:
            unique_lineups.add(datum)
    
    # Get DataFrames for each unique line (so we can use stats)
    # TODO: do this on the fly instead of storing all of them? If we have ID for each player we can just lookup then...
    lineup_dfs = []
    for lineup in unique_lineups:
        lineup_df = team.loc[list(lineup)]
        lineup_dfs.append(lineup_df)
    return lineup_dfs
        
def get_lineup_totals(lineups, week_data):
    """ Returns DF with each row being the total expected weekly stats for a lineup
    
        Params:
        - lineups = list of unique lineup dfs (returned from get_unique_lineups)
        - week_data = data for the week containing teams dict with number of games
        
        Returns:
        - Single DataFrame 
    """
    lineups_totals = []
    for i, line in enumerate(lineups):
        line_stats = line.filter(regex='_AVG')
        # Get number of games and multiple each row (player) by how many games they play
        line['NUM_GAMES'] = line['TEAM'].apply(lambda x: week_data['teams'][x])
        line[line_stats.columns] = line_stats.multiply(line['NUM_GAMES'], axis='index')

        line_total = Series(line.filter(regex='_AVG').sum(), name=i) #Sum each column across the whole line
        line_total['PLAYERS'] = line['PLAYER'].to_json(orient='values') # Store player list for easy reference later
        lineups_totals.append(line_total)
    return DataFrame(lineups_totals)

def get_absolute_wins(team_lineups):
    """ Calculate Absolute/guaranteed wins, 
            e.g. worst possible team 1 line for PTS beats best possible PTS team 2 line
        
        NOT BEING USED
    """
    team1_lineups = team_lineups[0]
    team2_lineups = team_lineups[1]
    team1_wins = (team1_lineups.min() > team2_lineups.max())
    team2_wins = (team2_lineups.min() > team1_lineups.max())
    team1_win_cats = team1_wins[team1_wins==True].index
    team2_win_cats = team2_wins[team2_wins==True].index
    return (team1_win_cats, team2_win_cats)

def get_top_lineups(lineups):
    """ Returns top 3 line ups for each team
        
        Params:
        - lineups = all unique lineups with total stats (returned from get_lineup_totals)
        
        Returns:
        - Tuple with a DF for each team with top 3 lines
        
        1. Compare team 1 line 1 to team 2 lines 1-N and totals the number of wins per line
        2. Sort by total expected wins (NOT FUCKING ACSENDING)
        3. Take top 3 lines
        
        TODO: This is by far the slowest function. Any way to not use two for loops?
    """
    teams_top_lines = []
    for i, team in enumerate(lineups):
        top_lines = []
        if i == 0:  #TODO: better way?
            other = 1
        else:
            other = 0
        for j,line1 in team.iterrows():
            lineup_wins = []
            line1_stats = line1.filter(regex='_AVG')
            for k,line2 in lineups[other].iterrows():
                line2_stats = line2.filter(regex='_AVG')
                lineup_wins.append((line1_stats > line2_stats))
            top_lines.append(DataFrame(lineup_wins).sum())
        
        top_lines = DataFrame(top_lines)
        top_lines['total'] = top_lines.sum(axis=1) # Sum wins across each row (lineup)
        top_lines = top_lines.sort(columns='total', ascending=False)[:3] # Take top 3
        top_lines = team.loc[top_lines.index] # Get Stats again for Top Lines (so we can compare later)
        teams_top_lines.append(top_lines)
    return teams_top_lines
    
def compare_top_lineups(top_lineups):
    """ Returns wins for top lines against eachother
    """
    wins = []
    for i,line1 in top_lineups[0].iterrows():
        line1_stats = line1.filter(regex='_AVG')
        for j,line2 in top_lineups[1].iterrows():
            line2_stats = line2.filter(regex='_AVG')
            wins.append((line1_stats>line2_stats))
    return wins

def calculate_match_lineups(match_teams, week_data):
    """ Gets the lineups and totals for each team
    """    
    team_lines = []
    for team_id in match_teams:
        team = players[players['OWNER_ID'] == team_id]
        team = team.query('not INJURED or (INJURED & not (INJURED_START <= @week_num < INJURED_END))')
        team_lineups = get_unique_lineups(team)
        lineups_totals = get_lineup_totals(team_lineups, week_data)
        team_lines.append(lineups_totals)
    return team_lines
```

```python
%%prun

""" Main cell that runs the stuff. 
"""

final_results = []
final_team_results = {}
for week in weekly_data:
    week_num = week['week']
    print ('week #{}'.format(week_num))
    week_results = {
        'week':week_num
    }
    for match in week['matches']:
        team1_owner = match['team1_owner']
        team2_owner = match['team2_owner']
        team1_id = match['team1_id']
        team2_id = match['team2_id']
        
        lineups = calculate_match_lineups((team1_id,team2_id),week)
        top_lines = get_top_lineups(lineups)
        team_results = compare_top_lineups(top_lines)
        
        median_wins = np.median([item.sum() for item in team_results])
        mode_wins = stats.mode([item.sum() for item in team_results], axis=None)[0][0]
        
        week_results[team1_owner + '_wins'] = mode_wins
        week_results[team2_owner + '_wins'] = 6-mode_wins
        
        if team1_owner not in final_team_results:
            final_team_results[team1_owner] = []
        if team2_owner not in final_team_results:
            final_team_results[team2_owner] = []
        final_team_results[team1_owner].append(team_results[0] + 0)
        final_team_results[team2_owner].append(1-team_results[0])
        
        #print('{} vs {}'.format(team1_owner, team2_owner))
        #print([item.sum() for item in team_results])
        if 'Joe Hand' in [team1_owner, team2_owner]:
            #print('{} vs {}'.format(team1_owner, team2_owner))
            #print([item.sum() for item in team_results])
            #print(top_lines[0])
            #print(top_lines[1])
            if 'Joe' in team1_owner:
                final_team_results[team1_owner][week_num-1]['OPP'] = team2_owner
                final_team_results[team1_owner][week_num-1]['TOP_LINE'] = top_lines[0].iloc[0]['PLAYERS']
            else:
                final_team_results[team2_owner][week_num-1]['OPP'] = team1_owner
                final_team_results[team1_owner][week_num-1]['TOP_LINE'] = top_lines[1].iloc[0]['PLAYERS']

        #break
    final_results.append(week_results)
    #break
```
