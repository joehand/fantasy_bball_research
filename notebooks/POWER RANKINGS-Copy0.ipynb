```python
PLAYER_FILE = '../Data/playerOwner_updated.csv'
FANTASY_SCHEDULE_FILE = '../Data/fantasy_schedule.csv'
NBA_SCHEDULE_FILE = '../Data/nbaFantasySched.csv'

players = pd.read_csv(PLAYER_FILE)
fantasy = pd.read_csv(FANTASY_SCHEDULE_FILE)
nba = pd.read_csv(NBA_SCHEDULE_FILE)

players['INJURED'] = players['INJURED'].fillna(False)
del(players['MIN_AVG'])
players['POS'] = players['POS'].apply(lambda x: literal_eval(x))
```

```python
weeks_data = []

NUM_WEEKS = 19

for week in range(1,NUM_WEEKS+1):
    #print (week)
    # Get schedule for NBA teams for this week (already adjusted for fantasy schedule)
    teams = nba[nba['Week'] == week].to_dict(outtype='records')[0]
    teams.pop("Week")
    teams.pop("Date")
    
    # Get matches and start/end dates
    #start_date = fantasy[fantasy['week'] == week]['start_date'].iloc(0)
    #end_date = fantasy[fantasy['week'] == week]['end_date'].iloc(0)
    matches = fantasy[fantasy['week'] == week][['team2_id', 'team2_owner', 'team1_id', 'team1_owner']].to_dict(outtype='records')
    
    item = {
        'week' : week,
        #'start_date' : start_date,
        #'end_date' : end_date,
        'teams' : teams,
        'matches' : matches
    }
    
    weeks_data.append(item)
```

```python
from itertools import product
from ast import literal_eval

LINEUP = [['PG'], ['SG'], ['SF'], ['PF'], ['C'],  ['PG', 'SG'], ['SF', 'PF', 'C'], ['PG', 'SG', 'SF', 'PF', 'C']]

def get_unique_lineups(team):
    player_positions = []
    for pos in LINEUP:
        player_pos = team[team['POS'].map(lambda x: any(i in pos for i in x))].index.values
        player_positions.append(player_pos.tolist())
    
    unique_lineups = set()
    for x in product(*player_positions):
        datum = frozenset(x)
        if len(datum) == 8:
            unique_lineups.add(datum)
    
    lineup_dfs = []
    for lineup in unique_lineups:
        lineup_df = team.loc[list(lineup)]
        lineup_dfs.append(lineup_df)
    return lineup_dfs
        
def get_lineup_totals(lineups, week_data):
    lineups_totals = []
    for i, line in enumerate(lineups):
        line['NUM_GAMES'] = line['TEAM'].apply(lambda x: week_data['teams'][x])
        line[line.filter(regex='_AVG').columns] = line.filter(regex='_AVG').multiply(line['NUM_GAMES'], axis='index')
        lineups_totals.append(Series(line.filter(regex='_AVG').sum(), name=i))
    return DataFrame(lineups_totals)

def get_absolute_wins(team_lineups):
    team1_lineups = team_lineups[0]
    team2_lineups = team_lineups[1]
    team1_wins = (team1_lineups.min() > team2_lineups.max())
    team2_wins = (team2_lineups.min() > team1_lineups.max())
    team1_win_cats = team1_wins[team1_wins==True].index
    team2_win_cats = team2_wins[team2_wins==True].index
    return (team1_win_cats, team2_win_cats)

def calculate_match_lineups(match_teams, week_data):
    """ match teams = tuple of TEAM_ID
    """    
    team_lines = []
    for team_id in match_teams:
        team = players[players['OWNER_ID'] == team_id]
        team = team.query('not INJURED or (INJURED & not (INJURED_START <= @week_num < INJURED_END))')
        team_lineups = get_unique_lineups(team)
        lineups_totals = get_lineup_totals(team_lineups, week_data)
        team_lines.append(lineups_totals)
    return team_lines

def get_top_lineups(lineups):
    teams_top_lines = []
    for i, team in enumerate(lineups):
        top_lines = []
        
        for j,line1 in team.iterrows():
            lineup_wins = []
            if i == 0:
                other = 1
            else:
                other = 0 #todo: better way?
            for k,line2 in lineups[other].iterrows():
                lineup_wins.append((line1 > line2))
            top_lines.append(DataFrame(lineup_wins).sum())
        
        top_lines = DataFrame(top_lines)
        top_lines['total'] = top_lines.sum(axis=1)
        top_lines = top_lines.sort(columns='total')[:3]
        top_lines = team.loc[top_lines.index]
        teams_top_lines.append(top_lines)
    return teams_top_lines
   
    
def compare_top_lineups(top_lineups):
    wins = []
    for i,line1 in top_lineups[0].iterrows():
        for j,line2 in top_lineups[1].iterrows():
            wins.append((line1>line2).sum())
    return wins
```
