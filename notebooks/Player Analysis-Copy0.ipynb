```python
%matplotlib inline
import matplotlib.pyplot as plt
 
import pandas as pd
from pandas import DataFrame

import seaborn as sns
```

```python
sns.set_style("ticks")
sns.set_context("talk")
```

```python
import mpld3
mpld3.enable_notebook()
```

```python
players = pd.read_csv('../Data/playerData.csv')
```

```python
NUM_PLAYERS = 150
```

```python
sets = ['_2014', '_ESPN', '_RW']

playDF = players

cols = playDF.filter(regex='_RW').columns
cols = [col for col in cols if not 'MIN' in col]

for col in cols:
    playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)

zscore_cols = 'RW_Zscore'
playDF['RW_Total_Zscore'] = playDF.filter(regex=zscore_cols).sum(axis=1)

playDF_Full = playDF

playDF = playDF.sort(columns=['RW_Total_Zscore'], ascending=False)
playDF = playDF[:NUM_PLAYERS]
playDF = playDF.drop(playDF.filter(regex='Zscore').columns, axis=1)

for item in sets:
    cols = playDF.filter(regex=item).columns
    
    cols = [col for col in cols if not 'MIN' in col]
    print(cols)
    
    for col in cols:
        playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)
    
    zscore_cols = item + '_Zscore'
    playDF[item[1:] + '_Total_Zscore'] = playDF.filter(regex=zscore_cols).sum(axis=1)
    zscore_cols = playDF.filter(regex=zscore_cols).columns
    zscore_cols = [col for col in zscore_cols if not 'AST' in col if not '3P' in col]
    playDF[item[1:] + '_Big_Zscore'] = playDF[zscore_cols].sum(axis=1)

playDF['RANK'] = playDF['RW_Total_Zscore'].rank(ascending=False)
playDF['POS_RANK'] = playDF.groupby('POS')['RW_Total_Zscore'].rank(ascending=False)
playDF = playDF.sort(columns=['RW_Total_Zscore', '2014_Total_Zscore'], ascending=False)
```

```python
df = playDF

TITLE='All Players Total Z-Score'

x = df['RANK']
y = df['RW_Total_Zscore']
#pos = df['POS']

fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))
scatter = ax.scatter(x, y, color='b', s=70, alpha=.6)
ax.grid(color='white', linestyle='solid', )
ax.set_title(TITLE)
ax.set_xlim(0,150)

labels = [player for player in df['PLAYER'].values]
tooltip = mpld3.plugins.PointLabelTooltip(scatter, labels=labels)
mpld3.plugins.connect(fig, tooltip)
```
