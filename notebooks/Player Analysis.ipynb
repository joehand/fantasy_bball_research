```python
%matplotlib inline
import matplotlib.pyplot as plt
 
import pandas as pd
from pandas import DataFrame

import seaborn as sns

import numpy as np
```

```python
sns.set_style("ticks")
sns.set_context("poster")
flatui = ["#34495e", "#3498db","#9b59b6", "#e74c3c", "#2ecc71", "#95a5a6",]
sns.palplot(sns.color_palette(flatui))
sns.set_palette(flatui)
```

```python
import requests
import io

SHEET_URL = 'https://docs.google.com/spreadsheet/ccc?key=19YpHBaqzp4uccm1q7BgT3ssC_U_TtPaP705xUsHIxN0&gid=1178982124&output=csv'

response = requests.get(SHEET_URL)
assert response.status_code == 200, 'Wrong status code'

draft = pd.read_csv(io.BytesIO(response.content))
draft['PRICE'] = draft['PRICE'].apply(lambda x: x.replace('$','')).astype(int)
draft['DRAFTED'] = True
draft.info()
```

```python
players = pd.read_csv('../Data/playerData_Avg.csv')
players = players.merge(draft, on='PLAYER', how='left')

NUM_PLAYERS = 150
sets = ['_2014', '_ESPN', '_RW', '_WIRE', '_AVG']

playDF = players[~players['PLAYER'].str.contains('Paul George')] # take out PG
playDF = playDF[~playDF['PLAYER'].str.contains('Metta World Peace')] # take out PG
playDF = playDF[~playDF['PLAYER'].str.contains('Andrew Bynum')] # take out PG
playDF['DRAFTED'] = playDF['DRAFTED'].fillna(False)

cols = playDF.filter(regex='_AVG').columns
cols = [col for col in cols if not 'MIN' in col if not 'PRICE' in col]

print (cols)
for col in cols:
    playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)

zscore_cols = 'AVG_Zscore'
playDF['AVG_Total_Zscore'] = playDF.filter(regex=zscore_cols).sum(axis=1)

playDF = playDF.sort(columns=['AVG_Total_Zscore'], ascending=False)
playDF = playDF[:NUM_PLAYERS]
playDF = playDF.drop(playDF.filter(regex='Zscore').columns, axis=1)

for item in sets:
    cols = playDF.filter(regex=item).columns
    
    cols = [col for col in cols if not 'MIN' in col if not 'PRICE' in col]
    
    print (cols)
    
    playDF = playDF[~(playDF[cols].T == 0).all()]
    for col in cols:
        playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)
    
    zscore_org_cols = item + '_Zscore'
    playDF[item[1:] + '_Total_Zscore'] = playDF.filter(regex=zscore_org_cols).sum(axis=1)

playDF['RANK'] = playDF['AVG_Total_Zscore'].rank(ascending=False)
playDF['POS_RANK'] = playDF.groupby('POS')['AVG_Total_Zscore'].rank(ascending=False)
playDF = playDF.sort(columns=['AVG_Total_Zscore', 'RW_Total_Zscore'], ascending=False)
playDF.info()
```

```python
TITLE='Top 150 players Total Z-Score (Average of RotoWorld, RotoWire, ESPN, RazzBall)'
X_VAR = 'RANK'
Y_VAR = 'AVG_Total_Zscore'
Y_VAR2 = '2014_Total_Zscore'
Y_ERR = 'RW_Total_Zscore|WIRE_Total_Zscore|ESPN_Total_Zscore|AVG_Total_Zscore'
GROUP = 'POS'
df = playDF.sort(Y_VAR, ascending=False)

css = """

tr.rank {
    border-bottom:1px solid #fff;
}
table.player
{
  border-collapse: collapse;
}
.player th
{
  color: #ffffff;
  background-color: #000000;
}
.player td
{
  background-color: #cccccc;
}
.player table,.player th,.player td
{
  font-family:Arial, Helvetica, sans-serif;
  border: 1px solid black;
  text-align: right;
}
"""

fig, ax = plt.subplots()
elements = []
labels=[]
for name, big_group in df.groupby(GROUP):
    positions = []
    color=next(ax._get_lines.color_cycle)
    for drafted, group in big_group.groupby('DRAFTED'):
        x = group[X_VAR]
        y = group[Y_VAR]
        y2 = group[Y_VAR2]
        yerr = [group[Y_VAR] - group.filter(regex=Y_ERR).min(axis=1), group.filter(regex=Y_ERR).max(axis=1) - group[Y_VAR]]
        
        if drafted:
            element = ax.scatter(x,y, color=color, 
                                 s=60, alpha=0.6, marker='D')
            element2 = ax.scatter(x,y2, color=color, 
                                 s=90, alpha=0.5, marker='+')
            ax.errorbar(x,y,yerr=yerr, linestyle="None", lw='1',color=color, )
            positions.append(element)
            positions.append(element2)
        else:
            element = ax.scatter(x,y, color=color, 
                                 s=90, alpha=0.2, marker='o')
            element2 = ax.scatter(x,y2, color=color, 
                                 s=90, alpha=0.5, marker='+')
            ax.errorbar(x,y,yerr=yerr, linestyle="None", lw='1',color=color, )
            positions.append(element)
            positions.append(element2)
        
        display_string = """ <table class="player">
                              <tr><th>{PLAYER}</th><th>({TEAM})</th></tr>
                              <tr class="rank"><th>{POS} Rank</th><td>{POS_RANK:1.0f}</td></tr>
                              <tr><th>MIN</th><td>{MIN_AVG}</td></tr>
                              <tr><th>PTS</th><td>{PTS_AVG}</td></tr>
                              <tr><th>REB</th><td>{REB_AVG}</td></tr>
                              <tr><th>AST</th><td>{AST_AVG}</td></tr>
                              <tr><th>STL</th><td>{STL_AVG}</td></tr>
                              <tr><th>BLK</th><td>{BLK_AVG}</td></tr>
                              <tr><th>3PM</th><td>{3PM_AVG}</td></tr>
                             </table>"""
        players = list(group.apply(lambda x: display_string.format(**x),axis=1))

        mpld3.plugins.connect(fig, mpld3.plugins.PointHTMLTooltip(element, players,
                                                       voffset=-25, hoffset=10, css=css))

    elements.append(positions)
    labels.append(name)

l1 = plt.Line2D([], [], color=sns.color_palette(flatui)[0], marker='D', lw=0)
l2 = plt.Line2D([], [], color=sns.color_palette(flatui)[0], marker='o', lw=0)
l4 = plt.Line2D([], [], color=sns.color_palette(flatui)[0], lw=1)
ax.legend([l1,l2,l4], ["Drafted", "Undrafted", "Proj. Range"],bbox_to_anchor=(0.25, 0.25),)   
ax.set_title(TITLE)
ax.set_xlim(0,(df[X_VAR].max() + 10))
mpld3.plugins.connect(fig, mpld3.plugins.InteractiveLegendPlugin(elements, labels))
mpld3.save_html(fig, '../../../../Dropbox/public/site/all_zscore.html')
```
