```python
%matplotlib inline

import pandas as pd
from pandas import DataFrame, Series

from scipy import stats
import numpy as np

from itertools import product
from ast import literal_eval
```

```python
""" Import the data. 

    We need three sets of data:
        - Player Database with stats, owners, positions, and injuries
        - Fantasy Schedule with team1 vs team2 (team IDs)
        - NBA Team schedule (according to fantasy weeks, e.g. combine all-star week into 1)
"""

PLAYER_FILE = '../Data/playerOwner_updated.csv'
INJURY_FILE = '../Data/playerOwner_inj.csv'
FANTASY_SCHEDULE_FILE = '../Data/fantasy_schedule.csv'
NBA_SCHEDULE_FILE = '../Data/nbaFantasySched.csv'

players = pd.read_csv(PLAYER_FILE)
players = players.merge(pd.read_csv(INJURY_FILE), how='left')
fantasy = pd.read_csv(FANTASY_SCHEDULE_FILE)
nba = pd.read_csv(NBA_SCHEDULE_FILE)

# Clean up some data
players['INJURED'] = players['INJURED'].fillna(False) 
players['INJURED_START'] = players['INJURED_START'].fillna(0) 
players['INJURED_END'] = players['INJURED_END'].fillna(0) 
del(players['MIN_2015']) # we don't need this for now
del(players['MIN_AVG']) # we don't need this for now
players['POS'] = players['POS'].apply(lambda x: literal_eval(x)) # make POS a list not string
```

```python
for col in players.filter(regex='_AVG').columns:
    name = col[:3]
    players[name + '_STATS'] = players[name + '_2015']
    #players[name + '_STATS'] = players[col]

    players[name + '_STATS'] = players[name + '_STATS'].fillna(value=players[col])
    #players[name + '_STATS'] = players[name + '_STATS'].fillna(value=players[name + '_2015'])
    
```

```python
class Team():
    POS_COL = 'POS'
    NBA_TEAM_COL = 'TEAM'
    STATS_COL_REGEX = '_STATS'
    BENCH_SPOTS = 2
    LINEUP = [['PG'], ['SG'], ['SF'], ['PF'], ['C'],  ['PG', 'SG'], ['SF', 'PF', 'C'], ['PG', 'SG', 'SF', 'PF', 'C']]

    def __init__(self, league, owner_id, owner=None):
        self.league = league
        self.NUM_WEEKS = league.NUM_WEEKS
        self.OWNER_ID_COL = league.OWNER_ID_COL
        self.TOTAL_ROSTER_SIZE = len(self.LINEUP) + self.BENCH_SPOTS
        
        self.id = owner_id
        if owner:
            self.owner = owner
        self.roster = league.playerDF[league.playerDF[self.OWNER_ID_COL] == owner_id]
        self.unique_lineups = self._create_unique_lineups()
        self.lineup_dfs = {}
        self.weekly_lines = {}
        self.predictions = []
     
    def __repr__(self):
        if self.owner:
            return self.owner
        else:
            return str(self.id)
    
    def _create_unique_lineups(self):
        """ Returns a list of each unique and valid lineup as DF

            Returns:
            - List of Unique Lineup DataFrames (5-45 lineups)

        TODO: This is pretty slow. 
              This only needs to be run once on a team, just need to figure out injuries ahead of time
        """
        roster_change_weeks = []
        if len(self.roster[self.roster['INJURED']] > 0):
            start_vals = self.roster.groupby('INJURED_START').size().index.values.astype(int)
            end_vals = self.roster.groupby('INJURED_END').size().index.values.astype(int)
            roster_change_weeks.extend(list(np.union1d(start_vals, end_vals)))
            
        if 0 not in roster_change_weeks:
            roster_change_weeks.append(0) 
        roster_change_weeks.sort()
        
        rosters = []
        for i, week_num in enumerate(roster_change_weeks):
            week_num = week_num+1
            if week_num >= self.NUM_WEEKS:
                continue
            if i+1 == len(roster_change_weeks):
                week_end = self.NUM_WEEKS
            else:
                week_end = roster_change_weeks[i+1] 
            roster = {
                'weeks_valid': (week_num,week_end)
            }

            team = self.roster.query('not INJURED or (INJURED & not (INJURED_START <= @week_num <= INJURED_END))')          
            if len(team) > self.TOTAL_ROSTER_SIZE:
                team = self._select_top_ten_players(team)
            
            # Create a list of POS posibilities, e.g. player_positions = [['Rondo', 'Smart'], ['Kawhi', 'Affalo'], ...]
            player_positions = []
            for pos in self.LINEUP:
                player_pos = team[team[self.POS_COL].map(lambda x: any(i in pos for i in x))].index.values
                player_positions.append(player_pos.tolist())

            # Combine all lists into a set of unique lineups of length 8
            unique_lineups = set()
            for x in product(*player_positions):
                datum = frozenset(x)
                if len(datum) == len(self.LINEUP):
                    unique_lineups.add(datum)
                    
            roster['unique_lineups'] = unique_lineups
            rosters.append(roster)
                
        return rosters
    
    def _select_top_ten_players(self, roster):
        """ selects top 10 players and returns roster DF
        """
        cols = roster.filter(regex='_AVG').columns
        for col in cols:
            roster[col + '_Zscore'] = (roster[col] - roster[col].mean())/roster[col].std(ddof=0)
        roster['Total_Zscore'] = roster.filter(regex='_Zscore').sum(axis=1)
        roster = roster.sort(columns=['Total_Zscore'], ascending=False)
        
        #print ('Dropping worst player: {}'.format(roster.iloc[10]['PLAYER']))
        return roster[:self.TOTAL_ROSTER_SIZE]    
    
    def get_lineup_data(self, week_num, index):
        """ Returns a DF for the specified lineup
        
            Stores as (week_num, index)
            
            TODO: This will unnecessarily duplicate lineups across weeks. How to make it better? 
        """
        if week_num > self.NUM_WEEKS:
            raise Exception('That be more weeks than we have.')
        
        if (week_num,index) in self.lineup_dfs:
            return self.lineup_dfs[(week_num,index)] 
        for group in self.unique_lineups:
            if group['weeks_valid'][0] <= week_num <=  group['weeks_valid'][1]:
                if index+1 > len(group['unique_lineups']):
                    raise Exception('index greater than number of lineups.')
                lineup = list(group['unique_lineups'])[index]
                lineup_df = self.roster.loc[list(lineup)]
                self.lineup_dfs[(week_num,index)] = lineup_df # store for quicker lookup later
                return lineup_df
            else:
                continue
    
    def calculate_lineup_totals(self, week_num, lineup_index=None, lineup_df=None):
        """ Returns Series with total expected weekly stats for a lineup
    
        Params:
        - week_num = week number
        - lineup_index = which lineup index we want to calculate for
        
        Returns:
        - Series 
        """
        if lineup_index is None and lineup_df is None:
            raise Exception('Need either DF or index')
        
        if lineup_index is not None:
            lineup_df = self.get_lineup_data(week_num, lineup_index)
        
        line_stats = lineup_df.filter(regex=self.STATS_COL_REGEX)
        lineup_df['NUM_GAMES'] = lineup_df[self.NBA_TEAM_COL].apply(lambda x: self.league.nba_sched[week_num][x])
        lineup_df[line_stats.columns] = line_stats.multiply(lineup_df['NUM_GAMES'], axis='index')
        line_total = Series(lineup_df.filter(regex=self.STATS_COL_REGEX).sum()) #Sum each column across the whole line
        line_total['PLAYERS'] = lineup_df['PLAYER'].to_json(orient='values') # Store player list for easy reference later
        return line_total
        
    def get_weekly_lineups(self, week_num):
        """ Returns unique_lineups for the week you want
            TODO: Should I return a whole DF here? Not sure what will be most usefull...
        """
        lineups = []
        for line_set in self.unique_lineups:
            valid_weeks = line_set['weeks_valid']
            if valid_weeks[0] <= week_num <= valid_weeks[1]:
                for i, line in enumerate(line_set['unique_lineups']):
                    lineups.append(self.calculate_lineup_totals(week_num, lineup_index=i))
                return DataFrame(lineups)
        return None
    
```

```python
class FantasyLeague():
    NUM_WEEKS = 19
    OWNER_COL = 'OWNER'
    OWNER_ID_COL = 'OWNER_ID'
    STATS_COL_REGEX = '_STATS'
    
    def __init__(self, nba_schedule, fantasy_schedule, players):        
        self.nba_sched = {(i+1):week for i,week in enumerate(nba_schedule.to_dict(outtype='records'))}
        
        self.matches = {}
        for week, group in fantasy_schedule.groupby('week'):
            self.matches[week] = group.to_dict(outtype='records')
        self.playerDF = players
        teams = list(players.groupby([self.OWNER_COL,self.OWNER_ID_COL]).size().index.values)
        self.teams = self._create_teams(teams)
        
    def _create_teams(self, teams):
        """ Create a team object for each owner. 
        """
        team_items = {}
        for team_owner, team_id in teams:
            team = Team(self, team_id, owner=team_owner)
            team_items[team_id] = team
        return team_items
    
    def get_top_lineups(self, week_number, teams):
        team_objs = (self.teams[teams[0]['id']], self.teams[teams[1]['id']])
        team_lines = (team_objs[0].get_weekly_lineups(week_number),
                      team_objs[1].get_weekly_lineups(week_number))
        team0_stats = team_lines[0].filter(regex=self.STATS_COL_REGEX)
        team1_stats = team_lines[1].filter(regex=self.STATS_COL_REGEX)
        team_lines[0]['wins'] = team0_stats.apply(lambda x: 
                                    team1_stats.apply(lambda y: x > y, axis=1).sum(), axis=1).sum(axis=1)
        team_lines[1]['wins'] = team1_stats.apply(lambda x: 
                                    team0_stats.apply(lambda y: x > y, axis=1).sum(), axis=1).sum(axis=1)
        
        top_lines = [team_line.sort('wins', ascending=False)[:3] for team_line in team_lines]
        
        for i,team in enumerate(team_objs):
            team.weekly_lines[week_number] = top_lines[i]
            
        return (top_lines[0], top_lines[1])
    
    def predict_matchup_outcome(self, top_weekly_lines):
        team_stats = [top_weekly_lines[0].filter(regex=self.STATS_COL_REGEX),top_weekly_lines[1].filter(regex=self.STATS_COL_REGEX)]
        
        team_lines = []
        for i,team in enumerate(top_weekly_lines):
            if i == 0:
                team['wins'] = team_stats[i].apply(lambda x: team_stats[1].apply(lambda y: x>y, axis=1).sum(), axis=1).sum(axis=1)
            else:
                team['wins'] = team_stats[i].apply(lambda x: team_stats[0].apply(lambda y: x>y, axis=1).sum(), axis=1).sum(axis=1)

            team_line = team.sort('wins', ascending=False).filter(regex=self.STATS_COL_REGEX)[:1]
            team_lines.append(team_line)
        
        try:  
            team0_results = DataFrame((team_lines[0].values > team_lines[1].values)+0, columns=team_lines[0].columns)
            team1_results = DataFrame((team_lines[1].values > team_lines[0].values)+0, columns=team_lines[1].columns)
        except:
            print(team_lines[0].values)
            print(team_lines[1].values)
            print(top_weekly_lines[1])
        
        return (team0_results, team1_results)
        
    def run_predictions(self, weeks=None):
        if not weeks:
            weeks = (1,self.NUM_WEEKS+1)
        prediction_results = []
        for week in range(weeks[0],weeks[1]):
            matches = self.matches[week]
            week_results = {'week':week}
            print("Predicting week: {}".format(week))
            for match in matches:
                teams = ({'id':match['team1_id'],'owner':match['team1_owner']},
                         {'id':match['team2_id'],'owner':match['team2_owner']})
                #print("Week {}, {} vs {}".format(week, teams[0]['owner'], teams[1]['owner']))
                #print(teams)
                match_results = self.predict_matchup_outcome(self.get_top_lineups(week, teams))
                """
                if teams[0]['id'] == 1:
                    print(self.get_top_lineups(week, teams)[0]['wins'].values)
                    print(self.get_top_lineups(week, teams)[0]['PLAYERS'].values)
                if teams[1]['id'] == 1:
                    print(self.get_top_lineups(week, teams)[1]['wins'].values)
                    print(self.get_top_lineups(week, teams)[1]['PLAYERS'].values)
                """
                for i in range(0,2):
                    results = match_results[i].iloc[0]
                    week_results[teams[i]['owner']] = results.sum(axis=1)
                    results['opponent'] = teams[i-1]['owner']
                    results['week'] = week
                    self.teams[teams[i]['id']].predictions.append(results)
            prediction_results.append(week_results)
        return DataFrame(prediction_results)
    
    
    """ Missing stuff
        - Better way to choose top lineups (nash equilibrium?)
        - Add current record to predictions (need to scrape that)
        - 
        
    """
```
