```python
players = pd.read_csv('../Data/playerData.csv')

NUM_PLAYERS = 150
sets = ['_2014', '_ESPN', '_RW', '_WIRE']

playDF = players[~players['PLAYER'].str.contains('Paul George')] # take out PG
playDF['DRAFTED'] = playDF['DRAFTED'].fillna(False)

cols = playDF.filter(regex='_RW').columns
cols = [col for col in cols if not 'MIN' in col]

for col in cols:
    playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)

zscore_cols = 'RW_Zscore'
playDF['RW_Total_Zscore'] = playDF.filter(regex=zscore_cols).sum(axis=1)

playDF_Full = playDF

playDF = playDF.sort(columns=['RW_Total_Zscore'], ascending=False)
playDF = playDF[:NUM_PLAYERS]
playDF = playDF.drop(playDF.filter(regex='Zscore').columns, axis=1)

for item in sets:
    cols = playDF.filter(regex=item).columns
    
    cols = [col for col in cols if not 'MIN' in col]
    
    playDF = playDF[~(playDF[cols].T == 0).all()]
    for col in cols:
        playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)
    
    zscore_org_cols = item + '_Zscore'
    playDF[item[1:] + '_Total_Zscore'] = playDF.filter(regex=zscore_org_cols).sum(axis=1)
    
    zscore_cols = playDF.filter(regex=zscore_org_cols).columns
    zscore_cols = [col for col in zscore_cols if not 'AST' in col]
    playDF[item[1:] + '_NOAST_Zscore'] = playDF[zscore_cols].sum(axis=1)
    
    zscore_cols = playDF.filter(regex=zscore_org_cols).columns
    zscore_cols = [col for col in zscore_cols if not '3P' in col]
    playDF[item[1:] + '_NO3_Zscore'] = playDF[zscore_cols].sum(axis=1)
    
    zscore_cols = playDF.filter(regex=zscore_org_cols).columns
    zscore_cols = [col for col in zscore_cols if not '3P' in col if not 'AST' in col]
    playDF[item[1:] + '_NO3AST_Zscore'] = playDF[zscore_cols].sum(axis=1)

playDF['RW_Total_Error'] = abs(playDF['RW_Total_Zscore'] - playDF['WIRE_Total_Zscore'])
playDF['RW_Total_Error1'] = abs(playDF['RW_Total_Zscore'] - playDF['ESPN_Total_Zscore'])
playDF['RW_Total_Error2'] = abs(playDF['ESPN_Total_Zscore'] - playDF['WIRE_Total_Zscore'])
playDF['RANK'] = playDF['RW_Total_Zscore'].rank(ascending=False)
playDF['NOAST_RANK'] = playDF['RW_NOAST_Zscore'].rank(ascending=False)
playDF['NO3_RANK'] = playDF['RW_NO3_Zscore'].rank(ascending=False)
playDF['NO3AST_RANK'] = playDF['RW_NO3AST_Zscore'].rank(ascending=False)
playDF['POS_RANK'] = playDF.groupby('POS')['RW_Total_Zscore'].rank(ascending=False)
playDF = playDF.sort(columns=['RW_Total_Zscore', '2014_Total_Zscore'], ascending=False)
playDF.info()
```
