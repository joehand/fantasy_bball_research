```python
sns.set_style("ticks")
sns.set_context("poster")
flatui = ["#34495e", "#3498db","#9b59b6", "#e74c3c", "#2ecc71", "#95a5a6",]
sns.palplot(sns.color_palette(flatui))
sns.set_palette(flatui)
```

```python
import mpld3
mpld3.enable_notebook()
```

```python
players = pd.read_csv('../Data/playerData.csv')
```

```python
NUM_PLAYERS = 150
```

```python
sets = ['_2014', '_ESPN', '_RW']

playDF = players[~players['PLAYER'].str.contains('Paul George')] # take out PG

cols = playDF.filter(regex='_RW').columns
cols = [col for col in cols if not 'MIN' in col]

for col in cols:
    playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)

zscore_cols = 'RW_Zscore'
playDF['RW_Total_Zscore'] = playDF.filter(regex=zscore_cols).sum(axis=1)

playDF_Full = playDF

playDF = playDF.sort(columns=['RW_Total_Zscore'], ascending=False)
playDF = playDF[:NUM_PLAYERS]
playDF = playDF.drop(playDF.filter(regex='Zscore').columns, axis=1)

for item in sets:
    cols = playDF.filter(regex=item).columns
    
    cols = [col for col in cols if not 'MIN' in col]
    
    playDF = playDF[~(playDF[cols].T == 0).all()]
    for col in cols:
        playDF[col + '_Zscore'] = (playDF[col] - playDF[col].mean())/playDF[col].std(ddof=0)
    
    zscore_org_cols = item + '_Zscore'
    playDF[item[1:] + '_Total_Zscore'] = playDF.filter(regex=zscore_org_cols).sum(axis=1)
    
    zscore_cols = playDF.filter(regex=zscore_org_cols).columns
    zscore_cols = [col for col in zscore_cols if not 'AST' in col]
    playDF[item[1:] + '_NOAST_Zscore'] = playDF[zscore_cols].sum(axis=1)
    
    zscore_cols = playDF.filter(regex=zscore_org_cols).columns
    zscore_cols = [col for col in zscore_cols if not '3P' in col]
    playDF[item[1:] + '_NO3_Zscore'] = playDF[zscore_cols].sum(axis=1)

playDF['RW_Total_Error'] = abs(playDF['RW_Total_Zscore'] - playDF['ESPN_Total_Zscore'])
playDF['RANK'] = playDF['RW_Total_Zscore'].rank(ascending=False)
playDF['NOAST_RANK'] = playDF['RW_NOAST_Zscore'].rank(ascending=False)
playDF['NO3_RANK'] = playDF['RW_NO3_Zscore'].rank(ascending=False)
playDF['POS_RANK'] = playDF.groupby('POS')['RW_Total_Zscore'].rank(ascending=False)
playDF = playDF.sort(columns=['RW_NOAST_Zscore','RW_Total_Zscore', '2014_Total_Zscore'], ascending=False)
playDF = playDF[playDF['POS'].str.contains('F|C')]
playDF.info()
```

```python
TITLE='BIG Men No-Ast Z-Score (RotoWorld)'
X_VAR = 'NOAST_RANK'
Y_VAR = 'RW_NOAST_Zscore'
Y_ERR = 'RW_Total_Error'
GROUP = 'POS'

css = """
.player
{
  padding:2px;
  color: #ffffff;
  background-color: #000000;
}
"""

df = playDF

fig, ax = plt.subplots()
elements = []
labels=[]
for name, group in df.groupby(GROUP):
    x = group[X_VAR]
    y = group[Y_VAR]
    yerr= group[Y_ERR]
    color=next(ax._get_lines.color_cycle)
    element = ax.scatter(x,y, color=color, 
                             s=90, alpha=0.2, marker='o')
    ax.errorbar(x,y,yerr=yerr, linestyle="None", lw='1',color=color, )
    elements.append([element])
    labels.append(name)
    
    players = ['<span class="player">{}</span>'.format(player) for player in group['PLAYER'].values]
    mpld3.plugins.connect(fig, mpld3.plugins.PointHTMLTooltip(element, players,
                                                   voffset=-15, hoffset=10, css=css))

ax.set_title(TITLE)
ax.set_xlim(0,(df[X_VAR].max() + 10))

mpld3.plugins.connect(fig, mpld3.plugins.InteractiveLegendPlugin(elements, labels))
mpld3.display()
```
