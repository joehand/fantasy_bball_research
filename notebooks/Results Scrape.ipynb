```python
import requests
import bs4
from itertools import islice


import pandas as pd
from pandas import DataFrame
```

```python
LOGIN_URL = 'https://r.espn.go.com/espn/memberservices/pc/login'
# Fill in your details here to be posted to the login form.
payload = {
    'username': 'filiwickers',
    'password': 'blueman',
    'submit': 'Sign In',
    'failedAttempts':2,
    'SUBMIT':1,
    'aff_code' : 'espn_fantgames',
    'cookieDomain' :'.go.com',
    'multipleDomains':'true',
    
}

SCOREBOARD_URL = 'http://games.espn.go.com/fba/scoreboard'

params = (
    ('leagueId',14029),
    ('matchupPeriodId', 1)
) 


results = []
TOTAL_WEEKS = 11
matchup_period = 1

with requests.Session() as s:
    s.post(LOGIN_URL, data=payload)
    while matchup_period <= (TOTAL_WEEKS):
        params = list(params)[:-1] + [('matchupPeriodId',matchup_period)]
        params = tuple(params)

        r = s.get(SCOREBOARD_URL, params=params)    
        print(r.url)
        
        soup = bs4.BeautifulSoup(r.text)
        results.extend(soup.find_all("tr", class_="tableBody"))
        matchup_period += 1    
    cols = soup.find("tr", class_="tableSubHead")
len(results)
```

```python
cols_clean = [col.text.strip() for col in cols.find_all('th')]
cols_clean.insert(1, '')
cols_clean.insert(8, '')
results_clean = [[col.text.strip() for col in week.find_all('td')] for week in results]
```

```python
df = DataFrame(results_clean, columns=cols_clean)

teams = pd.Series(df.groupby('NAME').size())
teams = teams.rename(index=lambda x: x.split('(')[0].strip())
teams = teams.to_dict()

for team in teams:
    teams[team] = {col:0 for col in cols_clean[2:8]}
    teams[team].update({col+'_AGAINST':0 for col in cols_clean[2:8]})

for match in zip(*[iter(results_clean)]*2):
    team_names = [team[0].split('(')[0].strip() for team in match]
    #print(team_names)
    for i, stat in enumerate(match[0][2:8]):
        i+=2
        #print(match[0][i], match[1][i])
        #print(int(match[0][i]) > int(match[1][i]))
        if int(match[0][i]) > int(match[1][i]):
            teams[team_names[0]][cols_clean[i]] += 1
            #print('TEAM: {}, STAT: {}, COUNT: {}'.format(
            #        team_names[0], cols_clean[i], teams[team_names[0]][cols_clean[i]]))
        elif match[0][i] < match[1][i]:
            teams[team_names[1]][cols_clean[i]] += 1
            #print('TEAM: {}, STAT: {}, COUNT: {}'.format(
            #       team_names[1], cols_clean[i], teams[team_names[1]][cols_clean[i]]))
        else:
            teams[team_names[0]][cols_clean[i]] += 0.5
            teams[team_names[1]][cols_clean[i]] += 0.5
        
        teams[team_names[0]][cols_clean[i] + '_AGAINST'] += int(match[1][i])
        teams[team_names[1]][cols_clean[i] + '_AGAINST'] += int(match[0][i])
```

```python
standings = DataFrame(teams).transpose()
standings.filter(regex='AGA').rank()
```
