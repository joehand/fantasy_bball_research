```python
players = pd.read_csv('../Data/player_zscore.csv')
```

```python
players.info()
```

```python
players['ADJ_Zscore'] = players['AVG_Total_Zscore'] + abs(players['AVG_Total_Zscore'].min())
players['$_Zscore'] = players['PRICE']/players['ADJ_Zscore']
players['ADJ_BIG_Zscore'] = players['AVG_Big_Zscore'] + abs(players['AVG_Big_Zscore'].min())
players['$_BIG_Zscore'] = players['PRICE']/players['ADJ_BIG_Zscore']
```

```python
ax = players.plot(y='ADJ_Zscore', x='PRICE', kind='scatter')
ax.set_xlim(ax.get_xlim()[::-1])
```

```python
ax = players.plot(y='ADJ_BIG_Zscore', x='PRICE', kind='scatter')
ax.set_xlim(ax.get_xlim()[::-1])
```
