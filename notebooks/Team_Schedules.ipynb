```python
import requests
import bs4

import pandas as pd
from pandas import DataFrame
```

```python
SCHED_URL = 'https://basketballmonster.com/ScheduleGrid.aspx'

with requests.Session() as s:
    r = s.get(SCHED_URL)    
    print(r.url)
    
    soup = bs4.BeautifulSoup(r.text)
  
    columns = soup.find('table', class_='gridT').find('tr', class_='gridHeaderTR')
    weeks = soup.find('table', class_='gridT').find_all('tr')
```

```python
cols = [col.text.strip() for col in columns.find_all('td')]
weeks = weeks[4:29] # throw away extra TR
weeks = [[col.text.strip() for col in week.find_all('td')] for week in weeks]
```

```python
df = DataFrame(weeks, columns=cols).set_index('Week')
del(df[''])
df
```

```python
df.to_json('../fantasy_app/app/data/teamSched.json', orient='records')
```
